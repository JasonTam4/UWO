### Question 1.2. Generate a new figure with an array of 6 subplots (2 rows and 3 columns recommended, see pages 220-223). In the subplots, provide a histogram for the variables age, educational-num, income, month of birth, hours per week, and total. For each variable, choose an appropriate number of bins (? / 4 points)

```python
#### getting the variables for which we want to plot the histograms
# for the discretely coded variables, calculate the difference between max and min

(fig, axes) = plt.subplots(nrows = 2, ncols = 3, figsize = (15, 10))
n_bin = df['age'].max()-df['age'].min()
axes[0, 0].hist(df['age'], bins = n_bin, color = 'b', alpha = 1)
axes[0, 0].set_xlabel('Age [year]')
axes[0, 0].set_ylabel('Count')

n_bin = df['educational-num'].max()-df['educational-num'].min()
axes[0, 1].hist(df['educational-num'], bins = n_bin, color = 'b', alpha = 1)
axes[0, 1].set_xlabel('education-num')
axes[0, 1].set_ylabel('Count')

axes[0, 2].hist(df['income'], bins = 50, color = 'b', alpha = 1)
axes[0, 2].set_xlabel('income')
axes[0, 2].set_ylabel('Count')

axes[1, 0].hist(df['month_of_birth'], bins = 12, color = 'b', alpha = 1)
axes[1, 0].set_xlabel('Month of birth')
axes[1, 0].set_ylabel('Count')

n_bin = df['hours-per-week'].max()-df['hours-per-week'].min()
axes[1, 1].hist(df['hours-per-week'], bins = n_bin, color = 'b', alpha = 1)
axes[1, 1].set_xlabel('Work hours per week')
axes[1, 1].set_ylabel('Count')

axes[1, 2].hist(df['total'], bins = 50, color = 'b', alpha = 1)
axes[1, 2].set_xlabel('Total')
axes[1, 2].set_ylabel('Count')

plt.show()
```

### Question 5.1.Using _groupby_ create two series, one that has the income for males, and one for females.  Use _groupby_ to first group the income values by gender and then use _get_group(genderValue)_ to create a new series for the specified gender (? / 4 points)

```python
genderIncome = df['income'].groupby(df['gender'])

#### group  income based on gender
maleIncome   = genderIncome.get_group('Male')
femaleIncome = genderIncome.get_group('Female')
```
